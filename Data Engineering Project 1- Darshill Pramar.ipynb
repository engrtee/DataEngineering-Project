### Loading the data into MySQL database
#### At this point i realized the dataset needed cleaning. Cleaning issues i observed were
#### 1. The account data table had two columns with an object datatype as against the integer data type i specified in the MySQL database
#### 2. It also had null values
#### 3. The Accounts Country data had duplicate values

```python
#To insert the data
accounts_data_table_insert = ("INSERT INTO AccountsData(CountryCode  , SeriesCode , CountryName , SeriesName , year_1995 , year_2000 , year_2005  , year_2010 , year_2015) VALUES ( %s, %s, %s, %s,%s,%s,%s,%s,%s);")
```

### Accounts Data Table Export

```python
#To check the data types of the columsn in the Account data dataframe
AccountsData.info()
```

#### This was solved by converting the columns to integer and the errors were coerced to ensure that all errors were converted to nan values which was later dropped.

```python
# To perform transformations on the columns. This would convert the columns to integer and the errors = coerce ensures all errors are converted to NAN values

AccountsData['1995 [YR1995]'] = pd.to_numeric(AccountsData['1995 [YR1995]'], errors='coerce') 
```

```python
AccountsData['2000 [YR2000]'] = pd.to_numeric(AccountsData['2000 [YR2000]'], errors='coerce') 
```

```python
#To ensure all nan values are dropped and the data set is clean and ready to be exported to the MySQL database
AccountsData=AccountsData.dropna(axis = 0)
AccountsData.head()
```

```python
#For loop to read the data from the dataframe into the sql database. iterrows is a function that iterates through every row in a pandas dataframe and the next line is running the insert query statement .
for i , row in AccountsData.iterrows():
    cur.execute(accounts_data_table_insert ,list(row))
```
